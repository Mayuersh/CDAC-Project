```python
from autoscraper import AutoScraper
```

```python
flipkart_url="https://www.flipkart.com/search?q=headphones&sid=0pm%2Cfcn%2C821%2Cfof&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=headphones%7CWired+Earphones&requestId=a3483ff1-9682-4bd2-bbbe-51bfe30d54c7"
wanted_list=["₹399","boAt BassHeads 100 Wired Headset"]
```

```python
scraper=AutoScraper()
result=scraper.build(flipkart_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(flipkart_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_6gbm':'Price','rule_eaga':'Price','rule_3mn8':'Price','rule_cz9e':'Price','rule_o4ai':'Title','rule_c94g':'Title'})
scraper.keep_rules(['rule_6gbm','rule_eaga','rule_3mn8','rule_cz9e','rule_o4ai','rule_c94g'])
scraper.save('flipkart-search')
```
